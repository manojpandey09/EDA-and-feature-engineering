| üî¢ Step | üìå Kya Check Karna Hai?                            | ‚úÖ Kya Karna Hai / Logic                                                                                       | üßÆ Formula / Code Example                                              | üß™ Output / Explanation                              |
|--------|---------------------------------------------------|----------------------------------------------------------------------------------------------------------------|------------------------------------------------------------------------|------------------------------------------------------|
| 1Ô∏è‚É£     | Data ka structure kya hai?                         | Rows, columns, datatypes, size check karo                                                                      | `df.shape`, `df.dtypes`, `df.info()`                                  | Dataset overview milta hai                        |
| 2Ô∏è‚É£     | Pehle kuch rows dekhni hai?                        | Top, bottom & random sample row check karo                                                                     | `df.head()`, `df.tail()`, `df.sample(5)`                              | Data kaisa dikhta hai, samajh aata hai            |
| 3Ô∏è‚É£     | Kya missing values hain?                           | Missing count aur % check karo                                                                                 | `df.isnull().sum()` <br> `(df.isnull().sum() / len(df)) * 100`         | Null value handle karne ke liye planning hoti hai |
| 4Ô∏è‚É£     | Descriptive stats chahiye?                         | Numerical & categorical columns ka summary nikalo                                                              | `df.describe()`, `df.describe(include='object')`                      | Mean, std dev, min, max milta hai                 |
| 5Ô∏è‚É£     | Column types segregate karne hain?                 | Numerical & Categorical alag karo                                                                              | `df.select_dtypes(include='number')` <br> `include='object'`           | Analysis simplify hota hai                       |
| 6Ô∏è‚É£     | Univariate analysis karni hai (1 column)?          | Numerical: hist, boxplot, skew <br> Categorical: value_counts, countplot                                       | `df['col'].hist()`, `sns.boxplot(x=df['col'])`, `df['col'].skew()`     | Distribution samajh aata hai                      |
| 7Ô∏è‚É£     | Outliers detect karne hain?                        | Boxplot, IQR method, Z-score use karo                                                                          | **IQR**: Q1, Q3 ‚Üí `Q3 - Q1` <br> **Z-score**: `zscore(df['col'])`      | Outlier handle karne me help karta hai            |
| 8Ô∏è‚É£     | Do column ka relation samajhna hai?                | Numerical vs Numerical: scatter, corr <br> Cat vs Num: boxplot <br> Cat vs Cat: crosstab                       | `sns.scatterplot()`, `sns.heatmap()`, `pd.crosstab()`                  | Feature relationships milte hain                  |
| 9Ô∏è‚É£     | Duplicate rows hain kya?                           | Duplicate count nikalo                                                                                         | `df.duplicated().sum()`                                               | Data cleaning me use hota hai                     |
| üîü     | Target column balanced hai kya?                    | Classification problems ke liye check karo                                                                     | `df['target'].value_counts(normalize=True)`                           | Imbalance detect hota hai                         |
| 1Ô∏è‚É£1Ô∏è‚É£  | Multicollinearity detect karni hai?                | Highly correlated features drop karo                                                                          | `sns.heatmap(df.corr())` <br> `drop columns with corr > 0.85`          | Redundant columns remove hote hain                |
| 1Ô∏è‚É£2Ô∏è‚É£  | Full EDA flow samajhna hai?                        | Step-wise flow follow karo                                                                                     | Summary below                                                         | Full process visualize hota hai                  |

# üìä Exploratory Data Analysis (EDA) ‚Äì Step-by-Step Ordered Breakdown

EDA ka main goal hai data ko samajhna, clean karna, aur model ke liye ready karna.
Neeche diya gaya hai **ordered structure of EDA**:

---

## 1Ô∏è‚É£ **Data Types & Structure Check**

| üîç **What**              | Check data type, structure, size              |
|--------------------------|-----------------------------------------------|
| üõ†Ô∏è Tools/Methods         | `df.info()`, `df.dtypes`, `df.shape`, `df.columns` |
| üéØ **Purpose**            | Pehle samjho ki kaunse columns numeric, categorical, ya datetime hain. Data size kitna bada hai. |

---

## 2Ô∏è‚É£ **Missing Value Analysis**

| üîç **What**              | Find missing/null values                      |
|--------------------------|-----------------------------------------------|
| üõ†Ô∏è Tools/Methods         | `df.isnull().sum()`, `sns.heatmap()`          |
| üéØ **Purpose**            | Samjho kitne columns me missing data hai, unka % kya hai, aur kya drop/impute karna hai. |

---

## 3Ô∏è‚É£ **Univariate Analysis (One Column)**

| üîç **What**              | Analyze one feature at a time                 |
|--------------------------|-----------------------------------------------|
| üõ†Ô∏è Tools/Methods         | `histplot`, `countplot`, `boxplot`, `describe()` |
| üéØ **Purpose**            | Column ka distribution, central tendency (mean/median), frequency ya outlier detect karna. |

---

## 4Ô∏è‚É£ **Outlier Detection**

| üîç **What**              | Find extreme or unusual values                |
|--------------------------|-----------------------------------------------|
| üõ†Ô∏è Tools/Methods         | `boxplot`, IQR method, Z-score                |
| üéØ **Purpose**            | Outliers modeling ko bigaad sakte hain. Visual ya mathematical methods se pakadna hota hai. |

---

## 5Ô∏è‚É£ **Bivariate Analysis (Two Columns)**

| üîç **What**              | Relationship between two variables            |
|--------------------------|-----------------------------------------------|
| üõ†Ô∏è Tools/Methods         | `scatterplot`, `boxplot(x, y)`, `barplot`, `crosstab` |
| üéØ **Purpose**            | Ek feature ka dusre ke sath kya relation hai (correlation, comparison, class balance) |

---

## 6Ô∏è‚É£ **Multivariate Analysis (3+ Columns)**

| üîç **What**              | Multiple features ka combined effect          |
|--------------------------|-----------------------------------------------|
| üõ†Ô∏è Tools/Methods         | `pairplot`, `catplot`, `FacetGrid`, `hue`     |
| üéØ **Purpose**            | Complex patterns aur clusters identify karna using more than 2 variables. |

---

## 7Ô∏è‚É£ **Correlation Analysis**

| üîç **What**              | Numerical features ke beech ka correlation     |
|--------------------------|-----------------------------------------------|
| üõ†Ô∏è Tools/Methods         | `df.corr()`, `sns.heatmap(annot=True)`        |
| üéØ **Purpose**            | Linear relationships aur feature selection ke liye important hai. |

---

## 8Ô∏è‚É£ **Target Variable Analysis**

| üîç **What**              | Target column ke sath relation analysis       |
|--------------------------|-----------------------------------------------|
| üõ†Ô∏è Tools/Methods         | `countplot`, `violinplot`, `boxplot`, groupby |
| üéØ **Purpose**            | Target ke imbalance ko check karna, aur features ka us par kya effect hai ye dekhna. |

---

## 9Ô∏è‚É£ **Time Series / Date-Based Analysis (If applicable)**

| üîç **What**              | Date/time data ke trend analyze karna         |
|--------------------------|-----------------------------------------------|
| üõ†Ô∏è Tools/Methods         | `lineplot`, `plot(x='date', y='value')`       |
| üéØ **Purpose**            | Seasonal trend, growth, decline ya pattern detect karna. |

---

## üîü **Final Visualization Summary**

| üîç **What**              | All important plots for story-telling         |
|--------------------------|-----------------------------------------------|
| üõ†Ô∏è Tools/Methods         | - Heatmaps<br>- Pairplots<br>- Countplots<br>- Boxplots<br>- Scatterplots |
| üéØ **Purpose**            | Data ko clearly dikhana for insights and stakeholder presentations. |

---

## üìå Bonus: Use These Plots As Per Situation

| üß™ Analysis Type       | üìà Plot Types                                             |
|------------------------|-----------------------------------------------------------|
| Univariate (1 col)     | `histplot`, `countplot`, `boxplot`, `kdeplot`            |
| Bivariate (2 cols)     | `scatterplot`, `boxplot`, `barplot`, `lineplot`          |
| Multivariate (3+ cols) | `pairplot`, `catplot`, `FacetGrid`, `hue-based` plots     |
| Correlation            | `heatmap` with `annot=True`                              |
| Time series            | `lineplot`, `.plot(x='date', y='value')`                 |

# Biverate analysis (2 coloumns )
-Bivariate analysis is the statistical analysis of exactly two variables to understand the relationship, association, or correlation between them.

| üß© Variable 1 (X) | üß© Variable 2 (Y)   | üß† Type of Analysis        | üìä Chart Type                | üõ†Ô∏è Function / Code Example                                         | üì¶ Library           |
| ----------------- | ------------------- | -------------------------- | ---------------------------- | ------------------------------------------------------------------- | -------------------- |
| **Categorical**   | **Categorical**     | Frequency/Association      | Grouped Bar Chart, Heatmap   | `pd.crosstab(df['X'], df['Y']).plot(kind='bar')`<br>`sns.heatmap()` | Pandas / Seaborn     |
| **Categorical**   | **Numerical**       | Distribution by Category   | Box Plot, Violin Plot        | `sns.boxplot(x='X', y='Y', data=df)`<br>`sns.violinplot()`          | Seaborn              |
| **Numerical**     | **Categorical**     | Same as above (just order) | Box Plot, Violin Plot        | `sns.boxplot(x='Y', y='X', data=df)`                                | Seaborn              |
| **Numerical**     | **Numerical**       | Correlation/Trend          | Scatter Plot, Line Plot      | `sns.scatterplot(x='X', y='Y', data=df)`<br>`sns.lineplot()`        | Seaborn / Matplotlib |
| **Numerical**     | **Numerical**       | Correlation Measurement    | Heatmap (Correlation Matrix) | `df.corr()`<br>`sns.heatmap(df.corr(), annot=True)`                 | Pandas / Seaborn     |
| **Categorical**   | **Target Variable** | Classification Trends      | Bar, Countplot, Cross-tab    | `sns.countplot(x='X', hue='target', data=df)`                       | Seaborn              |
